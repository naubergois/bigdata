## Exemplar topics  
Here are a dozen of exemplar topics (but this is a just as an example, you can/should come up with your own):

#### Pandas & Data Analysis
- Data collection, crawling, working with XML, JSON etc.
- Working with big DataFrames, Dask. 
- Optimizing memory usage while working with NumPy, Pandas and pure Python
- Working with JSON and XML in Pandas
- When to use SQL with Pandas or just SQL
- Feather data format
- Reducing memory consumption while analyzing data: tips & tricks
- Data analysis with bash (utilizing extremely efficient command line utils)
 
#### Visualization
 - Overview of Bokeh or another viz library
 - Working with geo-spatial data in Python
 - Creating animations for data analysis
 - Tree visualization in Python (smth more than standard graphviz)
 - Dimentionality reduction for visualization (there are some methods besides t-SNE, e.g. UMAP)
 
#### Decision trees & kNN
 - Decision trees with statistical tests in the nodes
 - Overview of H2O library
 - kNN in production systems: Annoy 
 - kNN as a strong baseline in recommender systems
 
#### Linear models

 - Poisson/quantile or another type of regression
 - Multi-label classification
 - Efficient implementation of linear models (based QR-decomposition or similar)
 - Median and quantile regression
 - Interpreting linear models: SHAP, eli5

#### Features & Validation
 - Counters in supervised learning tasks: WOE, smoothed likelihood and other methods of feature engineering based on the target feature
 - there is much more in Scikit-learn that we didn't cover (NestedCV, GroupKFold etc)

#### Bagging and Random Forest
 - Interpreting RF by reducing forests to trees
 - Compressing random forests
 - Various heuristics for assessing feature importance in forests and boosting
 
#### Unsupervised learning
 - Review of some clustering method (with motivation, why it is needed)
 - Efficient PCA implementation, Randomized PCA, online PCA
 - Unsupervised learning in real tasks
 - Word2Vec applied to sequential data (even website sessions like in the "Alice" competition)

#### SGD & Vowpal Wabbit
 - Something that better covers the course material, e.g. a broader review of Vowpal Wabbit 
 - Matrix factorization, FMs and Vowpal Wabbit implementation
 - FTRL algorithm: Follow the regularized leader

#### Time series
 - Predicting multiple time series at the same time
 - Detecting anomalies in time series
 
#### Boosting
 - CatBoost overview + examples where it works really well 
 - Overview of the H2O library
 - Gradient boosting and GPU speedup

#### Other
 - A/B tests and interleaving
 - Bayesian methods of hyperparameter optimization
 - Versioning datasets (ex. DVC.org)
 - Optimizing non-trivial metrics: tips & tricks
 - Closer to business metrics: uplift modelling (pylift)
